```python
import requests 
from bs4 import BeautifulSoup
```

```python
url = "http://www.rightmove.co.uk/house-prices/detail.html?country=england&locationIdentifier=REGION%5E1231&searchLocation=Southampton&radius=3.0&year=1"
r = requests.get(url)
c = r.content
soup = BeautifulSoup(c,"html.parser")
all = soup.find_all("div", {"class": "soldDetails"})
```

```python
page_num = 40
```

```python
list = []

base_url = url+"&index="
print(base_url)

for page in range(0,int(page_num)*25,25):
    print(base_url + str(page))
    r = requests.get(base_url + str(page))
    c = r.content
    soup = BeautifulSoup(c,"html.parser")
    all = soup.find_all("div", {"class": "soldDetails"})

    for item in all:
        d = {}


        #'address'
        try:
            d["Address"] = item.find("a", {"class":"soldAddress"}).text
        except:
            d["Address"] = None

        #'price'
        try:
            d["Price"] = item.find("td", {"class":"soldPrice"}).text[1:]
        except:
            d["Price"] = None

        #Building type
        try:
            d["Building Type"] = item.find("td", {"class":"soldType"}).text.split(",")[0]
        except:
            d["Building Type"] = None

        #Leasehold type
        try:
            d["Leasehold Type"] = item.find("td", {"class":"soldType"}).text.split(",")[1].strip()
        except:
            d["Leasehold Type"] = None

        #Number of Beds
        try:
            d["Bedrooms"] = item.find("td", {"class":"noBed"}).text.split(" ")[0]
        except:
            d["Bedrooms"]

        list.append(d)   
```

```python
import pandas as pd
```

```python
#function to remove commas
def no_commas(x):
    y = int(x.replace(',', ''))
    return y       
```

```python
df = pd.DataFrame(list)
df = df.dropna(subset=["Address"])
df['Price'] = df['Price'].apply(no_commas)
df
```

```python
def post(x):
    y = x[-8:-4]
    return y

def post_full(x):
    y = x[-8:]
    return y
    

df["Postcode"] = df["Address"].apply(post)
df["Postcode Full"] = df["Address"].apply(post_full)
df_pc = df.sort_values(by=['Postcode'])
df_pc.to_csv("Southampton_Property_PastYear.csv")
df_pc
```
